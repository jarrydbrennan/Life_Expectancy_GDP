```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import os
from google.colab import drive
drive.mount('/content/drive')
```

### Load file into dataframe & inspect

```python
file_path = '/content/drive/MyDrive/Codecademy/Life-Expectancy-and-GDP-Starter'
gdp = pd.read_csv(os.path.join(file_path, 'all_data.csv'))
print(gdp.head(10))
```

```python
gdp.info()
```

Has life expectancy increased over time in the six nations?

```python
life = 'Life expectancy at birth (years)'
sns.lmplot(x = 'Year', y = life, hue='Country', data = gdp, lowess = True)
plt.show()
plt.clf()
```

Has GDP increased over time in the six nations?

```python
sns.lmplot(x = 'Year', y = 'GDP', hue='Country', data = gdp, lowess = True)
plt.show()
plt.clf()
```

Is there a correlation between GDP and life expectancy of a country?

```python
sns.lmplot(x = life, y = 'GDP', hue='Country', data = gdp, lowess = True)
plt.show()
plt.clf()
```

What is the average life expectancy in these nations?

```python
sns.boxplot(x = 'Country', y=life, data = gdp, hue = 'Country')
plt.show()
plt.clf()
```

```python
life = 'Life expectancy at birth (years)'
print('From 2000 to 2015')
print('The average life expectancy in Chile: ', gdp[life][gdp.Country == 'Chile'].mean().round(2))
print('The average life expectancy in China: ', gdp[life][gdp.Country == 'China'].mean().round(2))
print('The average life expectancy in Germany: ', gdp[life][gdp.Country == 'Germany'].mean().round(2))
print('The average life expectancy in Mexico: ', gdp[life][gdp.Country == 'Mexico'].mean().round(2))
print('The average life expectancy in USA: ', gdp[life][gdp.Country == 'United States of America'].mean().round(2))
print('The average life expectancy in Zimbabwe: ', gdp[life][gdp.Country == 'Zimbabwe'].mean().round(2))
```

What is the distribution of that life expectancy?

```python
sns.histplot(gdp, x = life, hue = gdp.Country, bins = 50)
plt.show()
plt.clf()
```

From 2000 to 2015 life expectancy of the observed 6 nations has increased with Zimbabwe showing the greatest increase over the range. The other 5 nations had very similar rates of change in terms of life expectancy.

In that same time span, China saw the greatest rise in GDP, with the USA rising in a similar fashion but not as sharply.  Germany's rise began to platue around 2008, while Mexico and Chile showed very gradual growth over the time period. In comparison, Zimbabwe GDP nearly tripled from 2000 to 2015 yet barely appears to have risen due to the weightedness of China and USA's GDP.

There is a correlation between GDP and life expectancy. As GDP increase so does life expectancy. Countries with greater GDP experience longer lives, but the rate at whigh GDP rises does not correlate with the rate of increased life expectancy. Five of the six nations in this study average life expectancy is between 75 and 80 years, with Zimbabwe being an outlier at 50 years suggesting there are other factors a play in this African country not characterized by the other nations in this data set.
